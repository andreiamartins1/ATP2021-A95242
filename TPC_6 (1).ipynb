#### RESUMO:
Este trabalho, realizado na linguagem __python__, é um conjunto de algoritmos que resolve o problema proposto para TPC6: o desenvolvimento de uma janela de ilustração gráfica que permite visualizar os polinómios desenvolvidos no TPC5.

#### CÓDIGO:

```python
# -*- coding: utf-8 -*-
"""
Andeia Martins, A95242
TPC6

"""
import random
import matplotlib.pyplot as plt

def menu ():
    print (""" Bem vindo! As opções que se encontram disponíveis são:
           
(1): Criar um polinómio manualmente ou aleatoriamente;
(2) Visualizar o polinómio
(3): Ver o grau de um polinómio;
(4): Simplificar um polinómio;
(5): Calcular o polinómio num valor x;
(6): Tabela do polinómio numa margem de valores;
(7): 1ª derivada;
(8): 2ª derivada;
_______________________________________
(0) SAIR 
           """)


def criarPolinomio():
    opc=input ("Pretende criar o polinómio manualmente (M) ou aleatoriamente (A)? ")
    Pol= []
    if (opc=="M"):
        n= int (input("Quantos elementos tem o seu polinómio? "))
        for i in range (n):
            print ("Insira o coeficiente do ", i+1,"º elemento. ")
            coef=int (input())
            print("Insira o expoente do ", i+1,"º elemento. ")
            exp= int( input ())
            Pol.append((coef,exp))
        return Pol
    elif (opc=="A"):
        n= int (input("Quantos elementos quer que o seu polinómio tenha? "))
        for i in range (n):
            coef= random.randint(-100, 100)
            exp= random.randint(0, 100)
            Pol.append((coef,exp))
        return Pol
    
def verTermo(t):
    c, e = t
    if e==0:
        return (str (c))
    elif e==1:
        return(str(c)+ "x")
    else:
        return (str(c)+ "x^" + str(e))
    
def verPolinomio(p):
    res= "" 
    for t in p:
        if res=="":
            res= verTermo(t)
        else:
            res= res + " + " + verTermo(t)
    return (res)
    
def calcPol (p):
    x= int(input ("Quer calcular o polinómio para que valor x? "))
    y= 0
    for (c,e) in p:
        y= y + int( x**e * c)
    return y

def calcPoli (p,n):
    y= 0
    for (c,e) in p:
        y= y + int( n**e * c)
    return y

def grauPol (pol):
    exp= []
    grau=0
    for i in pol:
        c, e= i
        exp.append(e)
    for e in exp:
        if e > grau:
            grau=e
    return grau


def tabela (p):
    inferior= int(input ("Qual o menor valor de x? "))
    superior= int(input ("Qual o maior valor de x? "))
    print ("x" + " | " + "CalcPol")
    print ("-----------")
    for n in range (inferior, superior+1):
        print (str (n) + " | " + str (calcPoli(p, n)))

def keyPol (t):
    c,e= t
    return (e)

def ordenarPol (p):
    p.sort(reverse= True, key=keyPol) #decrescente
    return p
       
def simplificarPol (p):
    pord= ordenarPol(p)
    psimp=[]
    coef= []
    coeficiente=0
    n=1
    for i in pord: 
        while n<len(pord):
            c,e= i
            if (e == pord[n][1] ):
                coef.append(c)
            else:
                if len(coef)>0:
                    for i in coef:
                        coeficiente= coeficiente+i
                else:
                    coeficiente=c
            psimp.append((coeficiente, e))
            n=n+1
    return (psimp)

def derivarPol (p):
    res= []
    for (c,e) in p:
        if e > 0:
            res.append ((c*e, e-1))
    return res

def mostraPol(myPol):
    x=[i for i in range(-1000,1000)]
    y=[]
    for n in x:
        y.append(calcPoli(myPol,n))
        
    plt.plot(x,y)
    plt.plot(x, y, color='green')
    plt.xlabel('Abcissas')
    plt.ylabel('Ordenadas')
    plt.title("Representação gráfica do polinómio")
    plt.show()
    
op="8"
menu()
while op!="0":
    op= input ("Qual é a opção que pretende selecionar? ")
    if op=="1":
        pol= criarPolinomio()
        print (verPolinomio(pol))
    
    elif op=="2":
        try:
            mostraPol(pol)
        except:
            print ("Ainda não criou um polinómio. ")            
            
    elif op=="3":
        try:
            grau= grauPol(pol)
            print (grau)
        except:
            print ("Ainda não criou um polinómio. ")
        
    elif op=="4":
        poli= verPolinomio(simplificarPol(pol))
        print(poli)
        
    elif op=="5":
        try:
            resultado= calcPol(pol)
            print (resultado)
        except:
            print ("Ainda não criou um polinómio. ")
            
    elif op=="6":
        try:
            tab= tabela(pol)
        except:
            print ("Ainda não criou um polinómio. ")
            
    elif op=="7":
        try:
            derivada= verPolinomio(derivarPol(pol))
            print (derivada)
        except:
            print ("Ainda não criou um polinómio. ")
            
    elif op=="8":
        try:
            sderivada= verPolinomio(derivarPol(derivarPol(pol)))
            print (sderivada)
        except:
            print ("Ainda não criou um polinómio. ")
            
    elif op=="0":
        print ("Até à próxima! ")
    else:
        print ("Opção inválida! ")
```
