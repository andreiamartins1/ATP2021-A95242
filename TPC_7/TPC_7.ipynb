```python
import io
import matplotlib.pyplot as plt

#Converter
def getAluno(texto):
    textoLimpo2=texto.replace("\"","") 
    campos=textoLimpo2.split(",")
    aluno=[]
    aluno.append(campos[0]) #id
    aluno.append(campos[1]) #nome
    aluno.append(campos[2]) #curso
    aluno.append(campos[3:]) #notas
    lista=[]
    for s in aluno[3]:
        s=lista.append(int(s)) #converter de string para int
    aluno[3]=lista
    return aluno

#Leitura
def lerDataset(fnome):
    f=open(fnome,encoding="utf-8")
    bd=[]
    f.readline()
    for linha in f:
        bd.append(getAluno(linha))
    return bd

#Tabela
def chaveOrd(a):
    return a[1]

def listarDataset(bd):
    bd.sort(key=chaveOrd)
    print("id |        nome         curso      média")
    print("-----------------------------------------")
    for a in bd:
        print(a[0] + " | " + a[1] + " | " + a[2] + " | " + str(sum(a[3])/4))

#Consultar o registo 
def media(l):
    soma=0
    for i in range(0,len(l)):
        soma=soma+l[i]
    return soma/4


def consultarDataset(bd, id):
    for i in range(0,len(bd)):
        if bd[i][0]==id:
            med=media(bd[i][3])
            return print("INFO :: " + str(bd[i]) + "\nMEDIA :: " + str(med))
        else:
            if i==len(bd)-1:
                return print('Não existe nenhum aluno com esse id!')

#TOP10
def chaveOrdena(a):
    return media(a[3])

def top10(bd):                                   
    bd.sort(key=chaveOrdena, reverse= True)           #nova lista ordenada pelas medias e imprimir os 10 elementos da lista
    novaBD=[]                                        
    for aluno in range(0,10):                         
        novoAluno=[]                             
        for parametro in range(0,3):                 
            novoAluno.append(bd[aluno][parametro])  
        novoAluno.append(media(bd[aluno][3]))  
        novaBD.append(novoAluno)
    return novaBD

#Distribuições (curso e médias)
def distribPorCurso(bd):
    distribuicao ={}
    for a in bd:
        if a[2] in distribuicao.keys():
            distribuicao[a[2]]= distribuicao[a[2]] + 1
        else:
            distribuicao[a[2]]=1
    return distribuicao

def distribPorMedia(bd):
    distrib={}
    for a in bd:
        if str(round(media(a[3]))) in distrib.keys(): 
            distrib[str(round(media(a[3])))]= distrib[str(round(media(a[3])))]+1
        else:
            distrib[str(round(media(a[3])))]=1
    return(distrib)


#Gráficos
def plotDistribPorCurso(bd):
    c=distribPorCurso(bd)
    x=c.keys()
    y=[]
    for i in x:
        y.append(c[i])
    plt.bar(x, y, color='firebrick')
    plt.title("Gráfico de barras dos cursos")
    plt.show()
    

# EXERCICIO 8

def plotDistribPorMedia(bd):
   cm=distribPorMedia(bd)
   x=cm.keys()
   y=[]
   for i in x:
        y.append(cm[i])
   plt.bar(x, y, color='bisque')
   plt.title("Gráfico de barras das médias")
   plt.xlabel("Médias")
   plt.show()
```

```python
BD = []
BD = lerDataset("alunos.csv")  
print("EXERCÍCIO 4: ")
print("\n")
for linha in top10(BD):
    print(linha)
print("\n")
print("EXERCÍCIO 5: ")
print("\n")
print(distribPorCurso(BD))
print("\n")
print("EXERCÍCIO 6: ")
print("\n")
print(distribPorMedia(BD))
print("\n")
print("EXERCÍCIO 7: ")
print("\n")
plotDistribPorCurso(BD)
print("\n")
print("EXERCÍCIO 8: ")
print("\n")
plotDistribPorMedia(BD)    
```
